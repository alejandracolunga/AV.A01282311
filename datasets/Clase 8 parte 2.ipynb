```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

### 1. Existe alguna relación entre Brain Weight y Head Size?

```python
df = pd.read_csv('datasets/headbrain.csv')
df.head()

X = df['Head Size(cm^3)']
Y = df['Brain Weight(grams)']

#Coeficientes de correlación
np.corrcoef(X,Y)
```

```python
colors = {1: 'red', 2: 'green'}
plt.figure(figsize = (13,8))
plt.scatter(X,Y,
           c = df['Gender'].map(colors),
           marker = 'x',
           )
plt.xlabel('Head Size(cm^3)', fontweight = 'regular', color = 'black',
fontsize = 18, horizontalalignment = 'center')
plt.ylabel('Brain Weight(grams)', fontweight = 'regular', color = 'black',
fontsize = 18, horizontalalignment = 'center')

plt.xticks(fontsize = 18)
plt.yticks(fontsize = 18)
plt.title('Scatter plot de Head Size(cm^3) vs Brain Weight(grams)')
```

#### Conclusión
Por lo tanto, el coeficiente de correlacion es de 0.7995. Podemos decir que existe una relacion muy fuerte entre Brain Weight y Head Size.
