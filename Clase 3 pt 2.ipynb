### Pivot Tables in Python

```python
import pandas as pd
import numpy as np
```

```python
%ls datasets/
```

```python
df = pd.read_csv('datasets/census.csv')
df.head(2)
```

```python
my_pivot = pd.pivot_table(df,
                index = 'STNAME',
                          values = 'BIRTHS2010',
                          aggfunc = 'median',
                          columns = 'COUNTY',
                          margins = True
                )
my_pivot.head(2)
```

```python
df = pd.read_csv('datasets/sample_pivot.csv')
df.head(2)
```

```python
average_sales = pd.pivot_table(df,
                               index = 'Region',
                               values = 'Sales',
                               aggfunc = ['sum','mean','median'],
                               columns = 'Type',
                               margins = 'sum'
                              )
average_sales
```
