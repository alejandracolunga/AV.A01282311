```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('antropometria-dataset.csv')
df
```

```python
data = [df.loc[:,'cintura'],]

patches=plt.hist(data, bins = 15)
```

```python
sns.lmplot(x='sistol',
           y='cintura',
           data=df,
           fit_reg=False,
           hue= 'sexo'
          )
plt.title('Scatter plot of sistol vs cintura')
```

```python
sns.boxplot(data=df.loc[:,['cintura']])
```

```python
tabla = df.loc[df['sexo']==2,['sistol','diastol']]
tabla
```

```python
sns.lmplot(x = 'sistol',
          y = 'diastol',
          fit_reg = False,
          data = tabla)
```
