```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('antropometria-dataset.csv')
df
```

```python
table = pd.pivot_table(df, values=['peso'], index=['desc_ent'],
                    aggfunc={'peso': np.mean})
table
```

```python
table = pd.pivot_table(df, 
                       values=['peso'], 
                       index=['desc_ent'],
                       columns = ['sexo'],
                    aggfunc={'peso': np.mean})
table
```
