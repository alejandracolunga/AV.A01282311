```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
df = pd.read_csv('antropometria-dataset.csv')
df
```

```python
df.columns
```

```python
vars = ['diastol','sistol','edad', 'peso','cintura', 'cadera']
df = df[vars].dropna()
df
```

```python
df[vars].corr()
```

```python
model = sm.OLS.from_formula ('diastol ~ sistol + edad + peso + cintura',
                            data = df)
result = model.fit()
```

```python
y=result.resid
y.median()
```
